```python
# 분류2 

import os
import shutil
import pandas as pd
import numpy as np


original_path = '/home/workspace/data/.train/.task153/data/train'


df = pd.read_csv(os.path.join(original_path, 'train.csv'))

display(df)

train_df =  df.iloc[:int(14072*0.8),:]
validation_df = df.iloc[int(14072*0.8):,:]
train_df_0  = train_df.loc[train_df['label']==0, 'file_name'].values
train_df_1  = train_df.loc[train_df['label']==1, 'file_name'].values

validation_df_0 = validation_df.loc[validation_df['label']==0, 'file_name'].values 
validation_df_1 = validation_df.loc[validation_df['label']==1, 'file_name'].values


original_dataset    = '/home/workspace/data/.train/.task153/data/train'
base_dir = '/home/workspace/user-workspace/crack_data_full_8_2'

#=====================================폴더 생성 코드=====================================
if not os.path.exists(base_dir):
    os.mkdir(base_dir)        

train_dir = os.path.join(base_dir, 'train')
if not os.path.exists(train_dir):
    os.mkdir(train_dir)

validation_dir = os.path.join(base_dir, 'validation')
if not os.path.exists(validation_dir):
    os.mkdir(validation_dir)

test_dir = os.path.join(base_dir, 'test')
if not os.path.exists(test_dir):
    os.mkdir(test_dir)    

for j in range(2):
    train_class_dir = os.path.join(train_dir, '{}'.format(j)) 
    validation_class_dir = os.path.join(validation_dir,'{}'.format(j)) 
    if not os.path.exists(train_class_dir):
        os.mkdir(train_class_dir)
    if not os.path.exists(validation_class_dir):
        os.mkdir(validation_class_dir)
        
for path, direct, files in os.walk(os.path.join(original_dataset)):   
    for file in files:
        if file in train_df_0:
            src = os.path.join(original_dataset,file)
            dst = os.path.join(os.path.join(train_dir,'0'), file)
            shutil.copyfile(src, dst)                
        if file in train_df_1:
            src = os.path.join(original_dataset,file)
            dst = os.path.join(os.path.join(train_dir,'1'), file)
            shutil.copyfile(src, dst)                
        if file in validation_df_0:
            src = os.path.join(original_dataset,file)
            dst = os.path.join(os.path.join(validation_dir,'0'), file)
            shutil.copyfile(src, dst)                
        if file in validation_df_1:
            src = os.path.join(original_dataset,file)
            dst = os.path.join(os.path.join(validation_dir,'1'), file)
            shutil.copyfile(src, dst)
```
